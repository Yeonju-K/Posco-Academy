# 0. 패키지 불러오기 

```python
# 데이터 구성 : Series, DataFrame
import pandas as pd

# 행렬 연산 
import numpy as np

# 데이터 전처리 : 항목에 대한 index 생성 
from mlxtend.preprocessing import TransactionEncoder

# 지지도 계산
from mlxtend.frequent_patterns import apriori

# 연관 규칙 
from mlxtend.frequent_patterns import association_rules
```

# 1. 데이터 

### 1-1 데이터 생성

```python
df_raw=[["우유","라면"],
        ["라면"],
        ["아이스크림","우유"],
        ["과자","라면"],
        ["아이스크림"]]
df_raw
```

### 1-2 데이터 전처리

```python
# items 중 index 0번이 갖고 있으면 True, 아니면 False
enc=TransactionEncoder()
df_raw_enc=enc.fit_transform(X=df_raw)

# 위의 함수는 데이터를 행렬의 형태로 반환 pandas의 DataFrame의 형태로 변환
df_asso=pd.DataFrame(df_raw_enc,columns=enc.columns_)
df_asso.head()
```

# 2. 라이브러리 불러오기

### 2-1 규칙척도 임계값

```python
# 최소 지지도
v_min_sup=0.3

# 지지도 계산
df_freq=apriori(df_asso, min_support=v_min_sup, use_colnames=True)
df_freq.round(3)

# 소주가 장바구니 6개 중 2개에 들어있다. 
```

### 2-2 지지도, 신뢰도, 향상도 산출

```python
df_asso_rule=association_rules(df_freq, metric="lift", min_threshold=0)
df_asso_rule.round(3)
```

```python
df_raw=[["우유","라면"],
        ["라면"],
        ["아이스크림","우유"],
        ["과자","라면"],
        ["아이스크림"]]
df_raw
```

```python
# 고객 ID
ID=list(set(df_raw["ID"])) # 데이터를 가로로 변환 
ID.sort()

list_association=[]
for i in ID:
    # ID별 데이터 구매 추출 
    tmp_list=list(df_raw[df_raw["ID"]==i]["PRODUCT"])
    tmp_list.sort()
    # ID별 구매 데이터를 리스트에 담기 
    list_association.append(tmp_list)
    
for low in list_association:
    print(low)
```

```python
# items 중 index 0번이 갖고 있으면 True, 아니면 False
enc=TransactionEncoder()
df_raw_enc=enc.fit_transform(X=list_association)

# 위의 함수는 데이터를 행렬의 형태로 반환 pandas의 DataFrame의 형태로 변환
df_asso=pd.DataFrame(df_raw_enc,columns=enc.columns_)
df_asso
```

```python
# 최소 지지도
v_min_support= 0.3
# 지지도 계산
df_freq=apriori(df_asso, min_support=v_min_sup, use_colnames=True)
df_freq.round(3)
```

```python
df_asso_rule=association_rules(df_freq, metric="lift", min_threshold=1.5)
df_asso_rule.round(3)
```
