# 데이터 결합
#### 연결 : concat(A,B) : A와 B를 결합
#### 추가 : A.append B : A에 B를 결합

```python
# 데이터 생성  
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                index=[0,1,2,3])
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7'],
                  'D':['D4','D5','D6','D7']},
                      index=[4,5,6,7])
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                  'B':['B8','B9','B10','B11'],
                  'C':['C8','C9','C10','C11'],
                  'D':['D8','D9','D10','D11']},
                      index=[8,9,10,11])
```

```python
# concat을 이용한 데이터 결합 
result=pd.concat([df1,df2,df3])
result
```

```python
# axis=1 : 세로 결합이 아닌 가로 결합 
result=pd.concat([df1,df2,df3],axis=1)
result
```

```python
# ignore_index=True : 기존 index가 아닌 새로운 index로 재생성 
result1=pd.concat([df1,df2,df3], axis=1, ignore_index=True)
result1
```

```python
# concat 결과 데이터의 변수명 생성 
result1.columns=["col_"+str(i) for i in range(12)]
result1
```

# Merge로 합치기

```python
#left,right 데이터 형성 
left=pd.DataFrame({'key1':['K0','K1','K2','K3'],
                  'key2':['K0','K1','K0','K1'],
                  'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key1':['K0','K1','K2','K3'],
                  'key2':['K0','K0','K0','K0'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
# merge 결과 result 생성
result=pd.merge(left,right,on=['key1','key2'])
result
```

```python
#how=left 
result=pd.merge(left,right,on=['key1','key2'],how='left')
result
```

```python
#how=right
result=pd.merge(left,right,on=['key1','key2'],how='right')
result
```

```python
#how=outer 
result=pd.merge(left,right,on=['key1','key2'],how='outer')
result
```

```python
#indicator 
result=pd.merge(left,right,on=['key1','key2'],how='outer',indicator=True)
result
```

# join으로 합치기

```python
left=pd.DataFrame({'key1':[1,2,3,4],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'C':['C1','C2','C5'],
                    'D':['D1','D2','D5']},
                    index=[1,2,5])
```

```python
left
```

```python
right
```

```python
# left에 right join
result=left.join(right)
result
```

```python
# how=right 
result=left.join(right,how='right')
result
```

```python
# how=inner (교집합 but left 기반)
result=left.join(right,how='inner')
result
```

```python
# how=outer (합집합)
result=left.join(right,how='outer')
result
```

```python
# on : join의 key를 지정 (left의 column명)
result=left.join(right,on='key1') #조건이 없을 때는 index기준으로 
result
```

```python
# merge와 비교 
result2=pd.merge(left,right,left_on='key1',right_index=True, how='left') #key1를 기준으로 데이터 프레임을 구성하되 더할 때는 right의 index를 기준으로 더해준다. 
result2
```

# 서울시 구별 인구수와 CCTV 설치 비율 

```python
#데이터 불러오기
pop_seoul=pd.read_csv("서울인구통계.txt",sep="\t",engine="python",thousands=',',encoding="utf_8")
pop_seoul.head()
```

```python
#자치별 인구합계, 한국인합계, 외국인합계, 고령자 데이터만을 추출
pop1=pop_seoul[['자치구','인구','인구.3','인구.6','65세이상고령자']]

#불필요한 행 삭제
pop_seoul=pop1.drop([0,1])
pop_seoul.head()
```

```python
#컬럼명 구분할 수 있도록 리네임
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                         pop_seoul.columns[1]:'인구수',
                         pop_seoul.columns[2]:'한국인',
                         pop_seoul.columns[3]:'외국인',
                         pop_seoul.columns[4]:'고령자'},inplace=True)
pop_seoul.head()
```

```python
#데이터 불러오기
cctv_seoul=pd.read_excel('서울시CCTV.xlsx')
cctv_seoul.head()
```

```python
#인구 통계 자료와 통일을 하기 위해 기관명을 구별로 인덱스 이름 변경, inplace=True로 해야지 값까지 다 포함해서 가져감 
cctv_seoul.rename(columns={cctv_seoul.columns[0]:'구별'},inplace=True)
cctv_seoul.rename(columns={cctv_seoul.columns[1]:'cctv수'},inplace=True)
# sort_values:값을 기준으로 정렬 
cctv_seoul.sort_values(by='cctv수',ascending=True).head(10)
cctv_seoul.sort_values(by='cctv수',ascending=False).head(10)
```
